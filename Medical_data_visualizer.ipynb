```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Import data
df = pd.read_csv("medical_examination.csv")
```

```python
#Viewing Datafile
df
```

```python
#Add Column "overweight" which calculates the BMI of person
# BMI = weight(in kgs)/ (height(in mts)^2)
df["overweight"] = df["weight"]/(df["height"]**2)*10000
```

```python
df
```

```python
#If BMI is > 25 then the person is overweight. Use the value 0 for NOT overweight and the value 1 for overweight.
df['overweight'] = np.where(df["overweight"] > 25,1,0)

#another method for the same
# df.loc[~(df['overweight'] > 25), 'overweight'] = 0
# df.loc[df['overweight'] > 25, 'overweight'] = 1
```

```python
# Normalize the data by making 0 always good and 1 always bad. If the value of cholesterol is 1, make the value 0. If the value is more than 1, make the value 1.

df.loc[df["cholesterol"]==1 , "cholesterol"]=0
df.loc[df["cholesterol"]>1 , "cholesterol"]=1
```

```python
# Normalize the data by making 0 always good and 1 always bad. If the value of cholesterol is 1, make the value 0. If the value is more than 1, make the value 1.

df.loc[df["gluc"] == 1, "gluc"] = 0
df.loc[df["gluc"] > 1, "gluc"] = 1
```

```python
#draw heat map
def draw_heat_map():
    df_heat = df[(df['ap_lo']<=df['ap_hi']) & 
                 (df['height'] >= df['height'].quantile(0.025)) & 
                 (df['height'] <= df['height'].quantile(0.975)) &
                 (df['weight'] >= df['weight'].quantile(0.025)) &
                 (df['weight'] <= df['weight'].quantile(0.025))
                ]
    # Calculate the correlation matrix
    corr = df_heat.corr()
    
    # Generate a mask for the upper triangle
    mask = np.triu(corr)
    
    # Set up the matplotlib figure
    fig, ax = plt.subplots(figsize=(11, 9))
    
    # Draw the heatmap with 'sns.heatmap()'
    cmap = sns.diverging_palette(230, 20, as_cmap=True)
    ax = sns.heatmap(
        corr,
        center=0.04,
        annot=True,
        fmt='.1f',
        mask=mask)
    
    # Do not modify the next two lines
    fig.savefig('heatmap.png')
    return fig
```

```python
draw_heat_map()
```
